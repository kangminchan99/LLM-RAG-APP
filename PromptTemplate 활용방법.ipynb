```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

# temperature=0으로 설정하면 모델이 항상 가장 가능성이 높은 답변을 선택하도록 강제하여 일관된 출력을 제공합니다.
llm = ChatOpenAI(model="gpt-5.4", temperature=0)
question = "llm이 뭔가요?"
llm.invoke(question)
```

```python
text_prompt = f'아래 질문에 답변해주세요:\n\n {question}'
llm.invoke(text_prompt)
```

```python
# PromptTemplate
from langchain_core.prompts import PromptTemplate

prompt_with_template = '아래 질문에 답변해주세요:\n\n {question}'

# PromptTemplate을 사용할 경우 f-string이 아닌, 일반 스트링으로 하고 input_variables를
# 아래와 같이 작성해야한다.
prompt_template = PromptTemplate(template=prompt_with_template, input_variables=["question"])
prompt_template.invoke({"question": question})
```

```python
# prompt_template.invoke결과로 llm에 invoke 
llm.invoke(prompt_template.invoke({"question": question}))
```

```python
# 위의 작업을 chain을 활요하여 조금 더 간결하게 
prompt_chain = prompt_template | llm   # invoke한 결과가 알아서 들어간다
prompt_chain.invoke({"question": question})
```

```python
# RAG Pipeline 구현 - 웹 베이스 로더를 사용하여
%pip install -qU beautifulsoup4
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://en.wikipedia.org/wiki/Language_model")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1500, chunk_overlap=200)

document_list = loader.load_and_split(text_splitter)
```

```python
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings

embedding = OpenAIEmbeddings(model="text-embedding-3-large")

database = Chroma.from_documents(document_list, embedding, collection_name='wikipedia', persist_directory="./wikipedia_db")
```

```python
retriever = database.as_retriever(search_kwargs={"k": 1})
retriever.invoke(question)
retriever_docs = retriever.invoke(question)
retriever_docs
```

```python
# prompt template을 사용하여 변형시켜보기
rag_prompt_template = PromptTemplate(
  template= '''
  You are a helpful assistant for question answering about the wikipedia website.
  You are given the following context:
  {context}
  Question: {question}
  ''',
  input_variables=["context", "question"]
)
rag_chain = rag_prompt_template | llm
rag_chain.invoke({"context": retriever_docs, "question": question})
```

```python
# langchain rag-prompt 사용해보기
from langsmith import Client

client = Client()
rlm_rag_prompt = client.pull_prompt("rlm/rag-prompt")
```

```python
rlm_rag_chain = rlm_rag_prompt | llm
rlm_rag_chain.invoke({"context": retriever_docs, "question": question})
```

```python
# RunnablePassthrough는 입력을 그대로 출력하는 간단한 체인입니다.
from langchain_core.runnables import RunnablePassthrough

rag_chain = (
  {"context": retriever, "question": RunnablePassthrough()}
  | rlm_rag_prompt
  | llm
)
rag_chain.invoke(question)
```
